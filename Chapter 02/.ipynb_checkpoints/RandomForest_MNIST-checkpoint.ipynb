# This program does the multiclass classification of MNIST digits using Random Forest Classifier

The recognition of hand-written digits which is common problem in optical character recognition problem. This problem includes both locating and identifying the digit in the image. Below is the example code that uses Random Forest class of sklearn package to classify the digits to one of the 10 classes (0 to 9). The code gives the programming steps to achieve this and finally reports the evaluation metrics discussed above.

```python

from sklearn.datasets import load_digits
handw_digits = load_digits()
print (handw_digits.images.shape)

import matplotlib.pyplot as plt

```

```python
fig, axes = plt.subplots(1, 10, figsize=(8, 8),
                     subplot_kw={'xticks':[], 'yticks':[]},
                     gridspec_kw=dict(hspace=0.1, wspace=0.1))

for i, ax in enumerate(axes.flat):
    ax.imshow(handw_digits.images[i], cmap='binary', interpolation='nearest')
    ax.text(0.05, 0.05, str(handw_digits.target[i]),
            transform=ax.transAxes, color='green')

plt.show()

```

```python
X = handw_digits.data
print ("The input shape is: ", X.shape)

y = handw_digits.target
print ("The output shape is: ", y.shape)

```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

Xtrain, Xtest, ytrain, ytest = train_test_split(handw_digits.data, handw_digits.target,
                                    random_state=0)
print ("The training input data shape is: ", Xtrain.shape)
print ("The training target data shape is: ", ytrain.shape)
print ("The testing input data  shape is: ", Xtest.shape)
print ("The testing target data  shape is: ", ytest.shape)

```

```python
model = RandomForestClassifier(n_estimators=1000)
model.fit(Xtrain, ytrain)
print (model.get_params())
ypred = model.predict(Xtest)

import seaborn as sns; sns.set()

from sklearn.metrics import confusion_matrix
mat = confusion_matrix(ytest, ypred)
sns.heatmap(mat.T, square=False, annot=True, fmt='d', cbar=False)
plt.xlabel('target or actual label')
plt.ylabel('predicted label');
plt.show()
```
